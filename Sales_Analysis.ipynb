# Exploratry Data Analysis

## Gender

```python
ax = sns.countplot(x = 'Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen)
```

The above graph gives us the insight that the female purchases more than men and their purchasing power is also greater than men.

## Age

```python
df.columns
```

```python
ax = sns.countplot(x = 'Age Group' , hue = 'Gender' , data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group' , y = 'Amount' , data = sales_age)
```

The above graphs gives us the insight that the most buyers are female with age group of 26-35

## State

```python
sales_st = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x = 'State' , y = 'Orders' , data = sales_st)
```

```python
sales_st = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x = 'State' , y = 'Amount' , data = sales_st)
```

The above graphs gives us the insight that most of the orders and the total sales\amount is from Utar Pardesh, Maharashtra and Karnataka.

## Marital State

```python
ax = sns.countplot(x = "Marital_Status" , data = df)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_st = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'Marital_Status' , y = 'Amount' , hue = 'Gender' , data = sales_st)
```

Married women have done the most shopping and their purchasing power is also greater.

## Occupation

```python
ax = sns.countplot(x = 'Occupation' , data=df)

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x = 'Occupation' , y = 'Amount' , data = sales_occ)
```

Most people ordering are from IT sector, Healthcare and Aviation

## Product Category

```python
ax = sns.countplot(x = 'Product_Category' , data=df)

sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pc = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x = 'Product_Category' , y = 'Amount' , data = sales_pc)
```

Most sold products are from Food, clothing and Electronics category.

## Conclusion:

The given data gives us the insight that the most orders are being place by the maried women with age group 26-35 are working at IT sector, Healthcare and Aviation from the food, clothing and electronics category, respectively.
